
2.conversor de metros para milimetros.
Desenvolva um algoritimo que receba um valor em metros  e o converta para milimetros exibindo o resultado na tela

```python
metros = float(input('Digite o valor em metros: '))
milimetros = metros * 1000
print(f'{metros} metros equivalem a {milimetros} milímetros.')
```

3.calculadora de segundos

```python
dias = int(input('Digite a quantidade de dias: '))
horas = int(input('Digite a quantidade de horas: '))
minutos = int(input('Digite a quantidade de minutos: '))
segundos = int(input('Digite a quantidade de segundos: '))

total_segundos = (dias * 24 * 60 * 60) + (horas * 60 * 60) + (minutos * 60) + segundos

print(f'O total de segundos correspondente é: {total_segundos} segundos.')
```
